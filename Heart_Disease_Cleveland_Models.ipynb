# **Data Preparation**

## **Data Separation as X and Y**

```python
print(df.columns)
```

```python
y = df[' diagnosis']
y
```

```python
x = df.drop(' diagnosis', axis=1)
x
```

## **Data Splitting**

```python
from posixpath import split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

## **Dealing with '?' value**

```python
x_train = x_train.replace('?', pd.NA)
y_train = y_train.replace('?', pd.NA)

combined = pd.concat([x_train, y_train], axis=1) # Combine for aligned dropping
combined = combined.dropna() # Drop rows with missing values in either x_train or y_train
x_train = combined.iloc[:, :-1] # Extract x_train after dropping rows
y_train = combined.iloc[:, -1]  # Extract y_train after dropping rows


x_train = x_train.astype(float)
y_train = y_train.astype(float)
```

```python
x_test = x_test.replace('?', pd.NA)
y_test = y_test.replace('?', pd.NA)

combined = pd.concat([x_test, y_test], axis=1) # Combine for aligned dropping
combined = combined.dropna() # Drop rows with missing values in either x_train or y_train
x_test = combined.iloc[:, :-1] # Extract x_train after dropping rows
y_test = combined.iloc[:, -1]  # Extract y_train after dropping rows


x_test = x_test.astype(float)
y_test = y_test.astype(float)
```

# **Classification Models**

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import f1_score
```

#  **Model Builiding**

## **Random Forest**

### **Training the Model**

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=69)
rf.fit(x_train, y_train)
```

### **Applying the Model**


```python
pred_train_rf_y = rf.predict(x_train)
pred_test_rf_y = rf.predict(x_test)
```

### **Evaluating the Model**


```python
rf_train_accuracy = accuracy_score(y_train, pred_train_rf_y)
rf_train_corr = matthews_corrcoef(y_train, pred_train_rf_y)
rf_train_f1 = f1_score(y_train, pred_train_rf_y, average='weighted')

rf_test_accuracy = accuracy_score(y_test, pred_test_rf_y)
rf_test_corr = matthews_corrcoef(y_test, pred_test_rf_y)
rf_test_f1 = f1_score(y_test, pred_test_rf_y, average='weighted')
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_accuracy, rf_train_corr, rf_train_f1, rf_test_accuracy, rf_test_corr, rf_test_f1]).transpose()
rf_results.columns = ['Method', 'Training accuracy', 'Training correlation', 'Training F1', 'Test accuracy', 'Test correlation', 'Test F1']
rf_results
```
